```python
import json
import random
from faker import Faker #The most randomness of the json file

# Create a Faker instance
fake = Faker()

# Function to introduce randomness in data formatting
def randomize_data():
    # Random formats for names
    name = random.choice([
        fake.name(),             # Full name
        fake.first_name(),       # First name only
        fake.last_name(),        # Last name only
        fake.name().lower(),     # Full name in lowercase
        fake.first_name() + " "  # First name only
    ])

    # Random formats for dates
    dob = random.choice([
        fake.date_of_birth().strftime('%m/%d/%Y'),  # MM/DD/YYYY
        fake.date_of_birth().strftime('%Y-%m-%d'),  # YYYY-MM-DD
        fake.date_of_birth().isoformat(),           # ISO format
        str(fake.date_of_birth().year)              # Year only
    ])

    # Random formats for emails
    email = random.choice([
        fake.email(),                     # Complete email
        fake.user_name() + "@",           # Incomplete email
        fake.user_name() + "@example."    # Missing domain
    ])

    # Random formats for addresses
    address = random.choice([
        fake.address(),                    # Full address
        " ".join(fake.address().split()[:3]) + "...",  # Partial address
        fake.street_name()                 # Street name only
    ])

    # Other fields with some randomness
    phone = fake.phone_number() if random.random() > 0.2 else ""  # 20% chance of being empty
    ssn = fake.ssn() if random.random() > 0.1 else "XXX-XX-" + fake.ssn()[-4:]  # Partial SSN
    job_title = fake.job() if random.random() > 0.3 else fake.job().upper()  # Random case

    return {
        "name": name,
        "address": address,
        "email": email,
        "phone": phone,
        "ssn": ssn,
        "driver_license": fake.license_plate(),
        "passport_number": fake.bothify(text='????########', letters='ABCDEFGHIJKLMNOPQRSTUVWXYZ'),
        "credit_card": fake.credit_card_full(),
        "bank_account": fake.bban(),
        "company": fake.company(),
        "job_title": job_title,
        "username": fake.user_name(),
        "password": fake.password(),
        "url": fake.url(),
        "date_of_birth": dob,
        "medicine_name": fake.text(max_nb_chars=20),
        "disease_name": fake.text(max_nb_chars=20),
        "latitude_longitude": f"{fake.latitude()}, {fake.longitude()}",
        "random_text": fake.text(max_nb_chars=100),
        "isbn": fake.isbn13()
    }

# Generate multiple records
num_records = 30  # Adjust the number of records as needed
data = [randomize_data() for _ in range(num_records)]

# Output the data to a JSON file
with open('/Users/dkim1998/Downloads/fake_pii_data_very_dirty_validation.json', 'w') as f:
    json.dump(data, f, indent=4)

print("Unstructured fake PII data generated and saved to unstructured_fake_pii_data.json.")
```
